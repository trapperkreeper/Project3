```python
# Imports for loading environment variables.
import requests
import json
import pandas as pd
from dotenv import load_dotenv
import os
import time
```

```python
# Load environment variables.
load_dotenv()
# Store the API key in a variable.
MARKET_API = os.getenv("MARKET_API")

#dictionary to follow API documentation
params = {'access_key' : MARKET_API}
```

```python
date_from = '2015-01-01'
date_to = '2019-12-31'
limit = 1000
offset = 0
results_list = []
results_list_apple = []



url = "https://api.marketstack.com/v1/eod?symbols=AAPL"

while True:
    if offset == 2000:
        break
    query = f'{url}&access_key={MARKET_API}&date_from={date_from}&date_to={date_to}&offset={offset}&limit={limit}'
    print(query)
    results = requests.get(query).json()
    offset = offset+1000
    time.sleep(10)
    
    try:
        for doc in results["data"]:
            results_list_apple.append({
                "symbol":doc["symbol"],
                "date":doc["date"],
                "open":doc["open"],
                "close":doc["close"],
                "volume":doc["volume"],
                "high":doc["high"],
                "low":doc["low"],
                "split":doc["split_factor"],
                "dividend":doc["dividend"]
            })
            results_list.append({
                "symbol":doc["symbol"],
                "date":doc["date"],
                "open":doc["open"],
                "close":doc["close"],
                "volume":doc["volume"],
                "high":doc["high"],
                "low":doc["low"],
                "split":doc["split_factor"],
                "dividend":doc["dividend"]
            })
        print(f'{offset}')
        
    except:
        print("No records")
        False


```

```python
apple_df = pd.DataFrame(results_list_apple)
```

```python
apple_df.to_csv('data/aapl_stock_data.csv', index='False')
```
