```python
#dependancies needed for the charts

import plotly.graph_objects as go

import pandas as pd
from datetime import datetime
```

```python
def candlestick (value):
    chart = go.Figure(data=[go.Candlestick(x=value['date'],
                  open=value['open'],
                  high=value['high'],
                  low=value['low'],
                  close=value['close'])])
    return chart
```

```python
apple_stock_df = pd.read_csv('data/aapl_stock_data.csv')

#apple_stock_df.head()

chart = candlestick(apple_stock_df)
chart.show()
chart.write_image('data/apple_stock_candlestick.jpeg')
```
